```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

final_df = pd.DataFrame()

for j in range(1, 11):
    url = 'https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page={}'.format(j)
    headers = {"User-Agent": 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.81 Safari/537.36'}
    rep = requests.get(url, headers=headers).text
    soup = BeautifulSoup(rep, 'lxml')
    company = soup.find_all('div', class_='companyCardWrapper__primaryInformation')
    
    name = []
    rating = []
    service = []
    no_employees = []
    hq = []
    age = []

    for i in company:
        name.append(i.find('h2', class_='companyCardWrapper__companyName').text.strip())
        rating.append(i.find('span', class_='companyCardWrapper__companyRatingValue').text.strip())
        service.append(i.find_all('span', class_='companyCardWrapper__interLinking')[0].text.strip().split('|')[0])
        no_employees.append(i.find_all('span', class_='companyCardWrapper__interLinking')[0].text.strip().split('|')[1])
        hq.append(i.find_all('span', class_='companyCardWrapper__interLinking')[0].text.strip().split('|')[-1])
        age.append(i.find_all('span', class_='companyCardWrapper__interLinking')[0].text.strip().split('|')[-2])

    d = {'name': name, 'rating': rating, 'service': service, 'employees': no_employees, 'headquarter': hq, 'age': age}
    df = pd.DataFrame(d)

    final_df = pd.concat([final_df, df], ignore_index=True)

print(final_df.head())

```
